# Question 3: Relational DataFrame Simple Modelling

```python
# please run a logistic regression
# you would be predicting the variable "churn" using the variables:
# ["total day minutes","total night calls","customer service calls"] 

### your code here
import numpy as np
selected_features = df[['total day minutes', 'total night calls', 'customer service calls']]

from sklearn.model_selection import train_test_split
arrayT = np.array(df['churn'].values)
arrayF = np.array(selected_features.values)
X = arrayF
y = arrayT
X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.30, random_state=1)

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,Y_train)
y_pred=logistic_regression.predict(X_validation)
print('Accuracy: ',metrics.accuracy_score(Y_validation, y_pred))
```
